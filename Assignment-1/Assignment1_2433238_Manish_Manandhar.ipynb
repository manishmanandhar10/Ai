Tasks:

2. Define outliers using the 1.5 × IQR rule.

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/drive/MyDrive/Assignment1/south_asian_countries.csv')
q1 = data['score'].quantile(0.25)
q3 = data['score'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
outliers_score = data[(data['score'] < lower_bound) | (data['score'] > upper_bound)]
outliers_score
```

3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
file_path = pd.read_csv('/content/drive/MyDrive/Assignment1/south_asian_countries.csv')

# Define a function to detect outliers using the IQR rule
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Detect outliers for GDP per Capita and Score and add an outlier column
gdp_outliers = detect_outliers_iqr(file_path, 'Log GDP per capita')
score_outliers = detect_outliers_iqr(file_path, 'score')
file_path['Outlier'] = file_path.index.isin(gdp_outliers.index) | file_path.index.isin(score_outliers.index)

# Create the scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=file_path, x='Log GDP per capita', y='score',
                               hue='Outlier', palette={True: 'red', False: 'blue'}, s=100)

# Adjust legend
handles, labels = scatter_plot.get_legend_handles_labels()
scatter_plot.legend(handles=[handles[0], handles[1]], labels=['Normal', 'Outlier'], title='Indicator')

# Titles and labels
plt.title('Outlier Detection: GDP per Capita vs Score in South Asia')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.show()
```

4. Discuss the characteristics of these outliers and their potential impact on regional averages.

# **Characteristics of Outliers:**

**Extreme Values**: Outliers are data points that have unusually high or low values compared to the rest of the data. In this case, outlier countries in South Asia might have extremely high or low scores or GDP per capita.

**Deviation from the Trend**: Outliers deviate significantly from the general trend or pattern observed in the data. They may not follow the same relationship between GDP per capita and score as other countries in the region.


# **Potential Impact on Regional Averages:**

**Skewed Averages**: Outliers can disproportionately influence regional averages, pulling them up or down. This can lead to a misrepresentation of the typical or central tendency of the data for the region.

**Inflated Variability**: Outliers can increase the variability or spread of the data, making it appear more dispersed than it actually is. This can affect the interpretation of statistical measures like standard deviation.

3. Metric Comparisons:

between the regions using grouped bar charts.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data1 = pd.read_csv('/content/drive/MyDrive/Assignment1/south_asian_countries.csv')
data2 = pd.read_csv('/content/drive/MyDrive/Assignment1/middle_eastern_countries.csv')

south_asia_metrics = data1[['Country name', 'Log GDP per capita', 'Social support', 'Healthy life expectancy']]
middle_east_metrics = data2[['Country name', 'Log GDP per capita', 'Social support', 'Healthy life expectancy']]

south_asia_metrics['Region'] = 'South Asia'
middle_east_metrics['Region'] = 'Middle East'

combined_metrics = pd.concat([south_asia_metrics, middle_east_metrics], ignore_index=True)

avg_south_asia = south_asia_metrics[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()
avg_middle_east = middle_east_metrics[['Log GDP per capita', 'Social support', 'Healthy life expectancy']].mean()

metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy']
south_asia_values = avg_south_asia.values
middle_east_values = avg_middle_east.values

x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))

rects1 = ax.bar(x - width/2, south_asia_values, width, label='South Asia', color='green')
rects2 = ax.bar(x + width/2, middle_east_values, width, label='Middle East', color='blue')

ax.set_ylabel('Average Values')
ax.set_title('Comparison of Key Metrics Between South Asia and Middle East')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

plt.tight_layout()
plt.show()

disparity_gdp = abs(avg_south_asia['Log GDP per capita'] - avg_middle_east['Log GDP per capita'])
disparity_support = abs(avg_south_asia['Social support'] - avg_middle_east['Social support'])
disparity_health = abs(avg_south_asia['Healthy life expectancy'] - avg_middle_east['Healthy life expectancy'])

disparities = {
    'Log GDP per capita': disparity_gdp,
    'Social support': disparity_support,
    'Healthy life expectancy': disparity_health
}

largest_disparity_metric = max(disparities, key=disparities.get)
largest_disparity_value = disparities[largest_disparity_metric]

if avg_south_asia[largest_disparity_metric] > avg_middle_east[largest_disparity_metric]:
    largest_disparity_region = "South Asia"
else:
    largest_disparity_region = "Middle East"

print(f"The metric with the largest disparity is: {largest_disparity_metric}")
print(f"The largest disparity value is: {largest_disparity_value:.2f}")
print(f"The region with the larger value for {largest_disparity_metric} is: {largest_disparity_region}")
```

```python
import pandas as pd

data1 = pd.read_csv('/content/drive/MyDrive/Assignment1/south_asian_countries.csv')
data2 = pd.read_csv('/content/drive/MyDrive/Assignment1/middle_eastern_countries.csv')

south_asia_scores = data1['score']
south_asia_range = south_asia_scores.max() - south_asia_scores.min()

middle_east_scores = data2['score']
middle_east_range = middle_east_scores.max() - middle_east_scores.min()

print("Happiness Range Analysis:")
print("\nSouth Asia:")
print(f"Range (max - min): {south_asia_range:.2f}")

print("\nMiddle East:")
print(f"Range (max - min): {middle_east_range:.2f}")

if south_asia_range > middle_east_range:
    print("\nSouth Asia has a larger range in happiness scores.")
else:
    print("\nMiddle East has a larger range in happiness scores.")
```

6. Outlier Detection:
• Identify outlier countries in both regions based on Score and GDP per Capita.

• Plot these outliers and discuss their implications.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data1 = pd.read_csv('/content/drive/MyDrive/Assignment1/south_asian_countries.csv')
data2 = pd.read_csv('/content/drive/MyDrive/Assignment1/middle_eastern_countries.csv')

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]


south_asia_outliers_score = detect_outliers(data1, 'score')
south_asia_outliers_gdp = detect_outliers(data1, 'Log GDP per capita')

middle_east_outliers_score = detect_outliers(data2, 'score')
middle_east_outliers_gdp = detect_outliers(data2, 'Log GDP per capita')

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x=data1['Country name'], y=data1['score'], color='blue', label='Normal Data')
sns.scatterplot(x=south_asia_outliers_score['Country name'], y=south_asia_outliers_score['score'], color='red', label='Outliers')
plt.xticks(rotation=90)
plt.title('South Asia: Outliers in Happiness Score')
plt.xlabel('Country')
plt.ylabel('Happiness Score')

plt.subplot(1, 2, 2)
sns.scatterplot(x=data1['Country name'], y=data1['Log GDP per capita'], color='blue', label='Normal Data')
sns.scatterplot(x=south_asia_outliers_gdp['Country name'], y=south_asia_outliers_gdp['Log GDP per capita'], color='red', label='Outliers')
plt.xticks(rotation=90)
plt.title('South Asia: Outliers in GDP per Capita')
plt.xlabel('Country')
plt.ylabel('Log GDP per Capita')

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(x=data2['Country name'], y=data2['score'], color='blue', label='Normal Data')
sns.scatterplot(x=middle_east_outliers_score['Country name'], y=middle_east_outliers_score['score'], color='red', label='Outliers')
plt.xticks(rotation=90)
plt.title('Middle East: Outliers in Happiness Score')
plt.xlabel('Country')
plt.ylabel('Happiness Score')

plt.subplot(1, 2, 2)
sns.scatterplot(x=data2['Country name'], y=data2['Log GDP per capita'], color='blue', label='Normal Data')
sns.scatterplot(x=middle_east_outliers_gdp['Country name'], y=middle_east_outliers_gdp['Log GDP per capita'], color='red', label='Outliers')
plt.xticks(rotation=90)
plt.title('Middle East: Outliers in GDP per Capita')
plt.xlabel('Country')
plt.ylabel('Log GDP per Capita')

plt.tight_layout()
plt.show()

print("\nSouth Asia Outliers based on Happiness Score:")
print(south_asia_outliers_score[['Country name', 'score']])

print("\nSouth Asia Outliers based on GDP per Capita:")
print(south_asia_outliers_gdp[['Country name', 'Log GDP per capita']])

print("\nMiddle East Outliers based on Happiness Score:")
print(middle_east_outliers_score[['Country name', 'score']])

print("\nMiddle East Outliers based on GDP per Capita:")
print(middle_east_outliers_gdp[['Country name', 'Log GDP per capita']])
```

7. Visualization:
